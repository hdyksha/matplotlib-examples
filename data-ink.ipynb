```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
dat = [('Apple', 250), ('Orange', 400), ('Grape', 300), ('Lemon', 200), ('Peach', 350)]
labels, y = zip(*dat)
x = range(len(labels))
```

### Classic 

```python
colors = ['red', 'orange', 'blue', 'yellow', 'pink']
with plt.style.context('classic'):
    fig, ax = plt.subplots(figsize=(8, 6))
    
    for xx, yy, label, color in zip (x, y, labels, colors):
        ax.bar(xx, yy, width=0.5, label=label, color=color, align='center')
    
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.set_ylim(0, 450)
    ax.tick_params(labelsize=14)
    ax.yaxis.grid(color='gray', linestyle='solid')
    ax.set_axisbelow(True)

    ax.set_title('Classic', fontsize=18)
    ax.legend(bbox_to_anchor=(1.25, 0.6))
```

### High data-ink ratio

```python
gray = 'dimgray'
colors = [gray, gray, 'royalblue', gray, gray]

fig, ax = plt.subplots(figsize=(8, 6))

ax.bar(x, y, width=0.5, color=colors, edgecolor='none', align='center')

ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylim(0, 450)
ax.yaxis.grid(color=gray, linestyle='solid')
ax.set_axisbelow(True)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_color(gray)
ax.spines['bottom'].set_color(gray)
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')
ax.tick_params(direction='out', length=5, width=1.2, bottom='off', colors=gray, labelsize=14)
ax.set_title('Customized', color=gray, fontsize=18)

plt.show()
```
