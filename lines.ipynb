```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## Lines with simple color/linestyle/linewidth settings

```python
ymax = 4*np.pi #4π
x = np.linspace(0, ymax, 100)
funcs = [("sin", np.sin), ("cos", np.cos), ("myfunc1", lambda x: 2*x/ymax-1)]
colors = ['red', 'green', 'blue']
linestyles = ['solid', 'dashed', 'dashdot']
linewidths = [1, 3, 5]

fig, ax = plt.subplots()
for f, color, linestyle, linewidth in zip(funcs, colors, linestyles, linewidths):
    name, impl = f
    ax.plot(x, impl(x), label=name, color=color, linestyle=linestyle, linewidth=linewidth)

ax.legend(loc='lower right')
plt.show()
```

## Custom dashed lines

```python
ymax = 4*np.pi #4π
x = np.linspace(0, ymax, 100)
funcs = [("sin", np.sin), ("cos", np.cos), ("myfunc1", lambda x: 2*x/ymax-1)]
colors = ['red', 'green', 'blue']
# [100, 10, 30, 10]: 100 points on, 10 off, 30 on, 10 off
dashes_settings = [[100, 10], [100, 10, 30, 10], [30, 10, 100, 10]]

fig, ax = plt.subplots()
for f, color, linestyle, dashes in zip(funcs, colors, linestyles, dashes_settings):
    name, impl = f
    ax.plot(x, impl(x), label=name, color=color, linestyle='--', dashes=dashes)

ax.legend(loc='lower right')
plt.show()
```
